```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_vaccinations.csv')
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
df_india = df[df['country'] == 'India']
```

```python
df_india
```

```python
sns.set_style('darkgrid')
sns.lineplot(x='date',y='total_vaccinations',data=df_india)
```

```python
plt.figure(figsize=(20,10))
plt.tick_params(labelsize=10)
sns.barplot(y='date',x='total_vaccinations',data=df_india)
```

```python
plt.figure(figsize=(20,10))
plt.tick_params(labelsize=10)
sns.barplot(y='date',x='daily_vaccinations',data=df_india)
```

```python
df_test = df.drop_duplicates(subset ="country", keep = 'last', inplace = False)
```

```python
df_test
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(df_test.total_vaccinations, labels=df[pd.value_counts(df_test['total_vaccinations']).iloc[:10]]['country'], autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
plt.show()
```
